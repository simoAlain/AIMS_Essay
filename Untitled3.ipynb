### Setting the label for the classes used in this jupyter

```python
labels=['no_tumor','tumor']
```

### Setting the data for the modeling part.

```python
X_train = []
y_train = []
image_size = 128
for i in labels:
    folderPath = os.path.join('/content/drive/MyDrive/Brain_2_work','Training',i)
    for j in tqdm(os.listdir(folderPath)):
        img = cv2.imread(os.path.join(folderPath,j))
        img = cv2.resize(img,(image_size,image_size))
        X_train.append(img)
        y_train.append(i)
for i in labels:
    folderPath = os.path.join('/content/drive/MyDrive/Brain_2_work','Testing',i)
    for j in tqdm(os.listdir(folderPath)):
        img = cv2.imread(os.path.join(folderPath,j))
        img = cv2.resize(img,(image_size,image_size))
        X_train.append(img)
        y_train.append(i)
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
X_train, y_train = shuffle(X_train,y_train,random_state=101)
```

```python
X_train.shape
```

### Splitting the data to obtain the test and train set

```python
X_train,X_test,y_train,y_test=train_test_split(X_train,y_train,test_size=0.2)
```

```python
y_train_new=[]
for i in y_train:
    y_train_new.append(labels.index(i))
y_train = y_train_new
y_train = tf.keras.utils.to_categorical(y_train)

y_test_new=[]
for i in y_test:
    y_test_new.append(labels.index(i))
y_test = y_test_new
y_test = tf.keras.utils.to_categorical(y_test)
```

### Defining the CNN model used 

```python
def ConvBlock(model, layers, filters):
    '''Create [layers] layers consisting of zero padding, a convolution with [filters] 3x3 filters and batch normalization. Perform max pooling after the last layer.'''
   # dropout_rate = 0.2
    for i in range(layers):
        model.add(ZeroPadding2D((1, 1)))
        model.add(Conv2D(filters, (3, 3), activation='relu'))
        model.add(BatchNormalization(axis=3))
    model.add(MaxPooling2D((2, 2)))
   # model.add(Dropout(dropout_rate))
```

```python
def create_model():
    '''Create the FCN and return a keras model.'''

    model = keras.Sequential()

    # Input image: 75x75x3
    model.add(Lambda(lambda x: x, input_shape=(128, 128, 3)))
    ConvBlock(model, 1, 32)
    # 37x37x32
    ConvBlock(model, 1, 64)
    # 18x18x64
    ConvBlock(model, 1, 128)
    # 9x9x128
    ConvBlock(model, 1, 256)
    # 4x4x128
    model.add(ZeroPadding2D((1, 1)))
    model.add(Conv2D(2, (1, 1), activation='relu'))
    model.add(GlobalAveragePooling2D())
    # 4x4x2
    model.add(Activation('sigmoid'))
    
    return model
```

### Below, the model is tested for momentum and for Adam with different learning rate

```python
sgd = optimizers.SGD(lr=0.0001,momentum=0.99, nesterov=True)
```

```python
model=create_model()
```

```python
model.compile(loss="binary_crossentropy", optimizer=sgd, metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, validation_split=0.4, batch_size=16, epochs=250, initial_epoch=0)
```

```python
pred = model.predict(X_test)
pred = np.argmax(pred,axis=1)
y_test_new = np.argmax(y_test,axis=1)
```
